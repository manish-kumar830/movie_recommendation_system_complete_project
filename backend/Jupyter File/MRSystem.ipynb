```python
def convertIntoList(x):
  l=x.split('|')
  return l
```

```python
demo = movies
```

```python
demo['genres'] = demo['genres'].apply(convertIntoList)
```

```python
def joinInOne(x):
  l = []
  for i in x:
    l.append(i.replace(" ",""))
  return l
```

```python
demo['genres'] = demo['genres'].apply(joinInOne)
```

```python
demo['actors'] = demo['actors'].apply(convertIntoList)
```

```python
demo['actors'] = demo['actors'].apply(joinInOne)
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_movies_df['tags'] = new_movies_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
tag_vectors = cv.fit_transform(new_movies_df['tags']).toarray()
```

```python
tag_vectors
```

```python
tag_vectors[0]
```

```python
len(cv.get_feature_names_out())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(tag_vectors)
```

```python
similarity.shape
```

```python
x = sorted(list(enumerate(similarity[5])),reverse=True,key = lambda x: x[1])[1:6]
```

```python
x
```

```python
new_movies_df['original_title'][5]
```

```python
def recommend(movie_title):
  movie_index = new_movies_df[new_movies_df['original_title']==movie_title].index[0]
  similar_movies_list = sorted(list(enumerate(similarity[movie_index])),reverse=True,key = lambda x: x[1])[1:6]
  movies_dict  = {}
  index = 1
  for i in similar_movies_list:
    movies_dict[str(index)]={
        "imdb_id":new_movies_df.iloc[i[0]].imdb_id,
        "movie_name":new_movies_df.iloc[i[0]].original_title
    }
    index += 1
  print(movies_dict)
```

```python
recommend('Kaalia')
```
